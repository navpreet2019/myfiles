```python

#Monte Carlo Simulation and Efficient Frontier
#Applying Monte carlo simulation on portfolis using shape ratio
#creating efficient frontier based on Sharp Ratio
#MONTE CARLO SIMULATION

#Monte Carlo Simulation is a statistical method applied in financial modeling. ... The simulation relies on the
#repetition of random samples to achieve numerical results. It can be used to understand the effect of uncertainty and 
#randomness in forecasting models.

#Monte Carlo Simulation, also known as the Monte Carlo Method or a multiple probability simulation, is a mathematical
#technique, which is used to estimate the possible outcomes of an uncertain event.

import numpy as np
```

```python
#Monte Carlo Simulation is a great tool to master. It can be used to simulate risk and uncertainty that can affect 
#the outcome of different decision options.

#Simply said, if there are too many variables affecting the outcome, then it can simulate them and 
#find the optimal based on the values.

#It is a technique used to understand the impact of risk and uncertainty in prediction and forecasting models.
#You roll two dice. 

#When you roll 7 means 6 , then you gain 5 dollars.
#If you roll anything else than 7 means 6, you lose 1 dollar.
```

```python
#The efficient frontier is the set of optimal portfolios that offer the highest expected return for a defined level of risk
#or the lowest risk for a given level of expected return. 

#Portfolios that lie below the efficient frontier are sub-optimal
#because they do not provide enough return for the level of risk.
```

```python
# (1,7,2) 7 is exclusive means 6 here and taking it 2 times
# with monte carlo you are simulating randomness
def roll_dice():
    return np.sum(np.random.randint(1,7,2))
```

```python
roll_dice()
```

```python
#When you roll 7, then you gain 5 dollars.
#If you roll anything else than 7, you lose 1 dollar.
#results = np.zeros(2) means you are recording in results how many wins and how many losses 

def monte_carlo_simulation(runs=1000):
    results = np.zeros(2)
    for _ in range(runs):
        if roll_dice() == 7:
            results[0] += 1
        else:
            results[1] += 1
    return results
```

```python
# these are values of result[0] and result[1] for wins and losses
# we get 174*5 and lose 826*1
monte_carlo_simulation()
```

```python
# [0] for getting first entry
results=np.zeros(1000)
for i in range(1000):
    results[i]=monte_carlo_simulation()[0]
results
```

```python
#Monte Carlo Simulation with Portfolios and Sharpe Ratio for 
#getting the perfect balance of our stocks
import pandas_datareader as pdr
import datetime as dt
import pandas as pd
```

```python
tickers = ['AAPL', 'MSFT', 'TWTR', 'IBM']
start = dt.datetime(2020, 1, 1)

data = pdr.get_data_yahoo(tickers, start)
```

```python
data = data['Adj Close']
```

```python
data.head()
```

```python
log_returns = np.log(data/data.shift())
```

```python
log_returns
```

```python
weight = np.random.random(4)
weight /= weight.sum()
weight
```

```python
#expected return
exp_rtn = np.sum(log_returns.mean()*weight)*252
```

```python
exp_rtn
```

```python
#expected volatility
# to speed up using this formula
exp_vol = np.sqrt(np.dot(weight.T, np.dot(log_returns.cov()*252, weight)))
exp_vol
```

```python
# exp return/ expected volatility
sharpe_ratio = exp_rtn / exp_vol
sharpe_ratio
```

```python
# Monte Carlo Simulation
#runs=5000
n = 5000

weights = np.zeros((n, 4))
exp_rtns = np.zeros(n)
exp_vols = np.zeros(n)
sharpe_ratios = np.zeros(n)

for i in range(n):
    weight = np.random.random(4)
    weight /= weight.sum()
    weights[i] = weight
    
    exp_rtns[i] = np.sum(log_returns.mean()*weight)*252
    exp_vols[i] = np.sqrt(np.dot(weight.T, np.dot(log_returns.cov()*252, weight)))
    sharpe_ratios[i] = exp_rtns[i] / exp_vols[i]
```

```python
sharpe_ratios.max()
```

```python
sharpe_ratios.argmax()
```

```python
#apple and microsoft are high and twitter and ibm has low
weights[600]
```

```python
fig,ax=plt.subplots()
ax.scatter(exp_vols,exp_rtns,c=sharpe_ratios)
ax.scatter(exp_vols[sharpe_ratios.argmax()],exp_rtns[sharpe_ratios.argmax()],c='r')
ax.set_xlabel('Expected Volatility')
ax.set_ylabel('Expected Return')
```

```python

# estimating the stock beta via python-CAPM Regression
import pandas_datareader.data as reader
import datetime as dt
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
end=dt.datetime.now()
start=dt.date(end.year-5,end.month,end.day)
Input=['MSFT','^GSPC']
```

```python
df=reader.get_data_yahoo(Input,start,end)['Adj Close']
df
```

```python
# as we are intersted in monthly returns ,let's calculate them
#mtl_ret=df.resample('M').ffill().pct_change()

mtl_ret = df.resample('M').last().ffill().pct_change()
mtl_ret
# delete first row
mtl_ret=mtl_ret.dropna(axis=0)
mtl_ret.head()
```

```python
# building regression model
#define dependent and independent variable
X=mtl_ret['^GSPC']
y=mtl_ret['MSFT']
# adding intercept(by default not present)
X_sm=sm.add_constant(X)
```

```python
#building model using Ordinary Least Square Approach(OLS)
model=sm.OLS(y,X_sm)
results=model.fit()
results.summary()
```
